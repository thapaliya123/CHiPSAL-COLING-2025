```python
import gdown

def download_drive_file_by_id(file_id, target_file_name):
  dataset_url = f"https://drive.google.com/u/1/uc?id={file_id}&export=download"
  gdown.download(dataset_url, target_file_name)
  print("File Download succesfull")

FILE_ID = ['18q2d13h_7cV6FGdX3hnhrYCAKzMJzxhX', '1m6ZPxXS7YzLa2jXhoYQs0MOuVMEsg0vt']
FILENAME = ["train.csv", "val.csv"]
for file_id, file_name in zip(FILE_ID, FILENAME):
  download_drive_file_by_id(file_id, file_name)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

COL_NAMES = ['index', 'text', 'label']
train_df = pd.read_csv(FILENAME[0], header=0, names=COL_NAMES)
valid_df = pd.read_csv(FILENAME[1], header=0, names=COL_NAMES)
train_df.drop('index', axis=1, inplace=True)
valid_df.drop('index', axis=1, inplace=True)

train_df.head()
```

```python
# convert pandas dataframe to huggingface datasets library
from datasets import Dataset


train_dataset = Dataset.from_pandas(train_df)
valid_dataset = Dataset.from_pandas(valid_df)
print(train_dataset)
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("rahular/varta-bert")

def tokenize_function(examples):
  return tokenizer(examples['text'], padding="max_length", truncation=True, max_length=358)
  return tokenizer(examples['text'])

train_dataset = train_dataset.map(tokenize_function, batched=True)
valid_dataset = valid_dataset.map(tokenize_function, batched=True)
```

```python
from torch.utils.data import DataLoader

train_dataloader = DataLoader(train_dataset, shuffle=True, batch_size=8)
valid_dataloader = DataLoader(valid_dataset, batch_size=8)
```

```python
from transformers import AutoModelForSequenceClassification, AdamW

model_train = AutoModelForSequenceClassification.from_pretrained("rahular/varta-bert", num_labels=2)
model_train
```

```python
from transformers import Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir = "./results",
    num_train_epochs = 3,
    per_device_train_batch_size = 8,
    per_device_eval_batch_size = 8,
    learning_rate = 2e-5
)

trainer = Trainer(
    model = model_train,
    args = training_args,
    train_dataset = train_dataset,
    eval_dataset = valid_dataset
)

trainer.train()
```

```python
predictions = trainer.predict(valid_dataset)
print(predictions)
```

```python
import numpy as np

y_hat = np.argmax(predictions.predictions, axis=1)
y_true = valid_dataset['label']
print(y_hat)
print(y_true)
```

```python
from sklearn.metrics import classification_report, roc_auc_score

print(classification_report(y_true, y_hat))
```

```python
from scipy.special import softmax

y_pred_proba = softmax(predictions.predictions, axis=1)
y_pred_proba

print(roc_auc_score(y_true, y_pred_proba[:, 1]))
```
