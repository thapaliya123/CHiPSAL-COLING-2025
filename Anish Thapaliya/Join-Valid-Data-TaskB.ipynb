```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

VALID_TWEET_PATH = "/content/gdrive/MyDrive/NLP Competition/Dataset/Task B/SubTask-B-(index,tweet)val.csv"
VALID_LABEL_PATH = "/content/gdrive/MyDrive/NLP Competition/Dataset/Task B/SubTask-B-(index,label)val.csv"
JOINED_VALID_PATH = "/content/gdrive/MyDrive/NLP Competition/Dataset/Task B/SubTask-B-val.csv"

df_val_tweet = pd.read_csv(VALID_TWEET_PATH)
df_val_label = pd.read_csv(VALID_LABEL_PATH)

print(f"Tweet Df: {df_val_tweet.shape}")
print(f"Label Df: {df_val_label.shape}")
```

```python
## JOIN DATAFRAME
df_val = pd.merge(df_val_tweet, df_val_label, on='index')
unique_idx = df_val['index'].nunique()

print(f"Unique Index after Join: {unique_idx}")
print(f"DF shape after Join: {df_val.shape}")
```

```python
# Save DataFrame
df_val.to_csv(JOINED_VALID_PATH, index=False)
```
