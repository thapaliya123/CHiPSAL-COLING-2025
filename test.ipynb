```python
import pandas as pd


# Create DataFrame
df = pd.read_csv('ensemble_results/test.csv')

# Find the row with the maximum value
max_value_row = df.idxmax(axis=1)
max_value_row
```

```python
import pandas as pd
from sklearn.metrics import classification_report

# Load your test.csv with predicted labels (assuming it's already processed as df)
# df['Max_Class_Label'] contains your predicted labels

# Load valid.csv which contains the true labels
valid_df = pd.read_csv('data/taskc/valid.csv')  # Adjust path as needed

# Ensure that the valid.csv has the same number of rows as the predictions
# Assuming valid.csv has a 'True_Label' column with the actual labels
true_labels = valid_df['label'].apply(lambda x: int(x))

# Predicted labels are in df['Max_Class_Label']
predicted_labels = max_value_row.apply(lambda x: int(x))

# Create the classification report
report = classification_report(true_labels, predicted_labels, target_names=['Class 0', 'Class 1', 'Class 2'],digits=4)

print(report)
```
